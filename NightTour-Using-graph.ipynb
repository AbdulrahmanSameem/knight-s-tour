```python
import time

class Graph:
    
    def __init__(self):
        self.vertList = {}

    def addVertex(self,key):
        newVertex = Vertex(key)
        self.vertList[key] = newVertex
        return newVertex

    def has_key(self,n):
        return self.vertList.has_key(n)
    
    def addEdge(self,row,col,cost=0):
        if row not in self.vertList:
            nv = self.addVertex(row)
        if col not in self.vertList:
            nv = self.addVertex(col)
        self.vertList[row].add_Neighbor(self.vertList[col],cost)


class Vertex:
    def __init__(self, num):
        self.num = num
        self.adj = []
        self.visited = False
        self.cost = {}

    def add_Neighbor(self,nbr,cost=0):
        self.adj.append(nbr)
        self.cost[nbr] = cost
    def set_visited(self,visited):
        self.visited = visited
    def get_visited(self):
        return self.visited
    def getAdj(self):
        return self.adj
    def getNum(self):
        return self.num


def knightGraph(bdSize):

    def posToNodeId(r, c, s):
        return r*s+c

    ktGraph = Graph()
    # Build the graph

    for row in range(bdSize):
        for col in range(bdSize):
            nodeId = posToNodeId(row, col, bdSize)
            newPositions = genLegalMoves(row, col, bdSize)
            for e in newPositions:
                nid = posToNodeId(e[0], e[1], bdSize)
                ktGraph.addEdge(nodeId, nid)
    return ktGraph

def genLegalMoves(x, y, bdSize):
    newMoves = []
    Moves = [(-1,-2), (-1,2), (-2,-1), (-2,1),(1,-2), ( 1,2), ( 2,-1), ( 2,1)]
    for i in Moves:
        newRow = x + i[0]
        newCol = y + i[1]
        if legalCoord(newRow,newCol, bdSize):
            newMoves.append((newRow, newCol))
                 
    return newMoves

def legalCoord(row,col, bdSize):
    if row >= 0 and row < bdSize and col >= 0 and col < bdSize:
        return True
    else:
        return False

def dfs(path,vertix):
    if vertix.get_visited()==True:
        return
    vertix.set_visited(True)
    path.append(vertix)

    for vertix in vertix.getAdj():
        dfs(path,vertix)




size = int(input("Enter Matrix size: "))
start_time = time.time()
G = knightGraph(size)
path = []
dfs(path,G.vertList[0])

def nodeIdToPos(n):
    return (int(n/size), n%size)

def display(board):
    print()
    for x in range(size):
        for y in range(size):
            print (board[x][y],end=' ')
        print()
    print()

board = [[0 for x in range(size)] for y in range(size)]
x =0 
for i in range(size*size):
    s = path[i].getNum()
    r, c = nodeIdToPos(s)
    x+=1
    board[r][c] = x

display(board)
print("%s seconds" % (time.time() - start_time))
```
